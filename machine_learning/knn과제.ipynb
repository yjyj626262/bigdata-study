- 와인데이터 : 와인의 화학 성분을 사용하여 와인의 종류를 예측하기 위한 데이터셋

```python
from sklearn.datasets import load_wine

wine = load_wine()
print(wine.DESCR)
```

```python
wine.keys()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, random_state=42, shuffle=True)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_train, y_train))
print(knn.score(X_test, y_test))
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

acc = np.zeros([179])

for k in range(1, 134) :
    clf = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    yhat = clf.predict(X_test)
    acc[k-1] = accuracy_score(y_test, yhat)

max(acc)
```

```python
k = np.where(acc==0.7777777777777778)
k
```

```python
print(f'가장 정확도가 높은 k 값은 {k[0] + 1}입니다.')
```
