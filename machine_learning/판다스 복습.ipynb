## [ ] 연산자
- 데이터프레임에서 [] 안에 들어갈 수 있는 것은 컬럼명 문자, 또는 인덱스로 변환 가능한 표현식
- 컬럼지정 연산자로 이해하는 것이 좋음

```python
# 컬럼이 아닌 0을 넣는다면
titanic_df[0]
```

```python
# 하지만 인덱스로 변환 가능한 표현식을 사용한다면 가능
titanic_df[0:2]
```

```python
# 따라서 불린인덱스도 가능
titanic_df[titanic_df['Pclass'] == 3].head()
```

```python
# 정수형 슬라이싱 (iloc) : 불린인덱스는 제공하지 않는다.
titanic_df.iloc[1:3,[1,6,7]]
```

```python
# 마지막열
titanic_df.iloc[:,-1]
```

```python
# 마지막컬럼 제외
titanic_df.iloc[:,:-1].head()
```

```python
# loc[] 연산자
titanic_df.loc[4, 'Survived']
```

```python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head()
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
titanic_df.head()
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```
