```python
# 행렬 정렬
org_array = np.array([3,1,9,5])
np.sort(org_array)
```

```python
a = np.arange(15,-1,-1).reshape(4,4)
print(a)
print()
print(np.sort(a))
print()
print(np.sort(a, axis=0))
print()
print(np.sort(a, axis=1))
```

numpy에서 axis의 경우 pandas와는 반대다.

```python
# 행렬정렬 시에 원본 행렬의 인덱스를 반환해 주는 메소드인데 행렬 정렬은 진행하기 전에 적용하여 정렬을 진행했다는 가정하에 인덱스를 추출
print(org_array)
np.argsort(org_array)
```

```python
# 행렬 내림차순 정렬 시 원본 행렬 인덱스
np.argsort(org_array)[::-1]
```

```python
# 행렬 내적(곱)
a = np.arange(1,7).reshape(2,3)
b = np.arange(7,13).reshape(3,2)
np.dot(a, b)
```

```python
# 전치행렬
np.transpose(a)
```
